```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Lembrem-se sempre de alterar a importação dos dados de acordo com o diretório de vocês.
df = pd.read_csv("CHURN_TELECON_MOD08_TAREFA.csv", delimiter=';')

df.head(10)
```

Legenda dos dados:

*   **CustomerID** : Esse é o ID único de cada cliente nosso da rede de produtos de telecomunicação.

*   **Genero** : O genero dos nossos clientes.

*   **Idoso** : Se nosso cliente tem acima de 60 anos ou não. (Sim = 1 | Não = 0)

*   **Casado** : Se nosso cliente é casado ou não.

*   **Dependents** : Se cliente tem dependentes.

*   **Tempo_como_Cliente** : Tempo em meses que nosso cliente está com a nossa empresa.

*   **PhoneService** : Se ele contratou serviço telefônico com a nossa empresa.

*   **Servico_Internet** : Qual o tipo de serviço de internet contratado.

*   **Servico_Seguranca** : Se ele contratou serviço de segurança com a nossa empresa.

*   **Suporte_Tecnico** : Se já acionou o suporte técnico.

*   **StreamingTV** : Se ele contratou serviço de streaming para TV.

*   **Tipo_Contrato** : Qual o tipo de contrato dos nossos clientes, se era mensal, anual etc.

*   **Tipo_Contrato** : Qual o tipo de contrato dos nossos clientes, se era mensal, anual etc.

*   **PaymentMethod** : Forma de Pagamento.

*   **Pagamento_Mensal** : Quanto o cliente pagava mensalmente.

*   **Total_Pago** : Total já pago pelo cliente para nossa companhia.

*   **Churn** : Nossa coluna mais importante, se nosso cliente nos abandonou ou não.




# 1 - Carregue a base, verifique se os tipos de dados estão de acordo com cada coluna e caso não estejam realize a transformação dos tipos de dados.

```python
# 1. Verificando o estado atual dos tipos de dados e o uso de memória
print("--- Tipos de Dados e Uso de Memória (ANTES da Otimização) ---")
display(df.info())

# 2. Transformação de dados: Otimizando os tipos (Nível Avançado)
# Lista de colunas que representam categorias limitadas
colunas_categoricas = ['Genero', 'Casado', 'Dependents', 'PhoneService', 
                       'Servico_Internet', 'Servico_Seguranca', 'Suporte_Tecnico', 
                       'StreamingTV', 'Tipo_Contrato', 'PaymentMethod', 'Churn']

# Convertendo colunas do tipo 'object' (strings) para 'category'
for col in colunas_categoricas:
    df[col] = df[col].astype('category')

# Otimizando as colunas numéricas (Downcasting para economizar memória)
# 'Idoso' possui apenas valores 0 e 1, logo int8 (8 bits) é suficiente
df['Idoso'] = df['Idoso'].astype('int8')

# 'Tempo_como_Cliente' atinge no máximo em torno de 72 meses, logo int16 é perfeitamente adequado
df['Tempo_como_Cliente'] = df['Tempo_como_Cliente'].astype('int16')

# 3. Verificando o resultado e a redução no impacto na memória
print("\n--- Tipos de Dados e Uso de Memória (DEPOIS da Otimização) ---")
display(df.info())
# obs: comentários adicionados pós análise
```

# 2 - Esse exercício faremos por etapas:
A) Verifique se temos colunas com dados faltantes e traga a % dos dados faltantes por coluna.

```python
# A) Verificando % de dados faltantes por coluna
percentual_faltantes = (df.isnull().sum() / len(df)) * 100

print("--- Percentual de Dados Faltantes (%) ---")
# Filtramos para exibir apenas colunas que possuem dados ausentes, ordenando de forma decrescente
display(percentual_faltantes[percentual_faltantes > 0].sort_values(ascending=False))
```

B) Exclua nessa etapa as linhas das colunas com dados faltantes que você analisar que esse método é o ideal. Justifique o motivo de estar excluindo essas linhas.

```python
# Justificativa para exclusão de LINHAS:
# 1. Churn (0.20% faltantes): Sendo a NOSSA VARIÁVEL ALVO (Target), nós nunca devemos imputá-la. 
#    Fazer isso significa "inventar" a resposta para o modelo treinar, o que gera um viés gravíssimo.
# 2. Genero (0.48% faltantes): O percentual de perda é tão insignificante (< 1%) que compensa 
#    mais simplesmente descartar as linhas afetadas do que realizar uma imputação e possivelmente distorcer os dados.

# Removendo apenas as linhas onde 'Churn' ou 'Genero' são nulos
df.dropna(subset=['Churn', 'Genero'], inplace=True)

print("Dimensões do dataset após a remoção:", df.shape)
```

C) Nessa etapa substitua pela média, mediana ou moda as linhas das colunas com valores faltantes que você analisou e entendeu que esse era o melhor método. Não se esqueça de justificar sua escolha.

**Dica variáveis numéricas:** Não se esqueça de verificar a distrbuição dos dados, média e mediana.

**Dica variáveis categóricas:** Caso precise substituir uma variável categórica utilize um gráfico de barras para verificar a que mais aparece e utilizar ela na substituição.
Ou você pode optar por excluir a coluna, mas justifique sua escolha.

Caso opte por substituir você pode utilizar a função fillna:
df['SUACOLUNA'].fillna("VALOR QUE SUBSTITUIRA O NULO", inplace=True)
Ela funcionará como a replace.

```python
# Justificativa para as ações em C:
# 1. PhoneService (Aprox 60% faltante): Quando a perda de dados é tão massiva, qualquer tentativa de imputação 
#    irá criar um pico artificial enorme e destruir a distribuição original. 
#    A prática ideal em Data Science neste cenário é EXCLUIR A COLUNA inteira.
# 2. Pagamento_Mensal (Aprox 13% faltante): Como é uma variável numérica contínua, usamos a MEDIANA.
#    Optamos pela mediana pois ela é uma medida de tendência central muito mais robusta contra 
#    valores discrepantes (outliers) do que a média.

# 1. Excluindo a coluna PhoneService que está irremediavelmente comprometida
df.drop(columns=['PhoneService'], inplace=True)

# 2. Preenchendo Pagamento_Mensal com a Mediana
mediana_pagamento = df['Pagamento_Mensal'].median()

# Usando o padrão moderno do Pandas 2.0+ (Copy-on-Write) sem o uso do inplace em fatias:
df['Pagamento_Mensal'] = df['Pagamento_Mensal'].fillna(mediana_pagamento)

# Verificação final para mostrar que nosso Dataset está limpo!
print("\nVerificação Final - Total de Nulos restantes no Dataset:")
display(df.isnull().sum().sum())
```

# 3 - Verifique se encontramos valores digitados incorretamente, ou com letras maiusculas ou minusculas, ou até mesmo mesmos valores porém escritos de forma dirente e corrija.

```python
# 3) Verificação e correção de valores digitados incorretamente

print("--- Valores com Erros de Digitação (Antes da Correção) ---")
display(df['Genero'].unique())
display(df['Servico_Internet'].unique())

# Corrigindo 'Genero' e 'Servico_Internet'
# Dica de ouro: Sempre converta para object antes de usar o replace em categorias para evitar erros no Pandas
df['Genero'] = df['Genero'].astype('object').replace({'F': 'Female', 'f': 'Female', 'M': 'Male'}).astype('category')
df['Servico_Internet'] = df['Servico_Internet'].astype('object').replace({'dsl': 'DSL'}).astype('category')

print("\n--- Valores Corrigidos (Depois) ---")
display(df['Genero'].unique())
display(df['Servico_Internet'].unique())
```

# [EXTRA] 4 - Caso deseje, para manter um padrão na sua base, renomeie as colunas ou deixando em português, ou inglês.

```python
# [EXTRA] 4) Padronização e Tradução Completa do Dataset!

# 1. Traduzindo os nomes das Colunas restantes que estão em Inglês
dicionario_colunas = {
    'customerID': 'ID_Cliente',
    'Dependents': 'Dependentes',
    'PaymentMethod': 'Metodo_Pagamento',
    # Nota: A palavra "Churn" é um jargão universal da área de Dados no Brasil.
    # Optamos por mantê-la como 'Churn', mas se preferir, pode ser 'Evasao'.
}
df.rename(columns=dicionario_colunas, inplace=True)

# 2. Traduzindo as Categorias dentro das colunas! (Nível Avançado)
traducao_geral = {
    'Yes': 'Sim', 
    'No': 'Não',
    'No internet service': 'Sem internet',
    'Female': 'Feminino',
    'Male': 'Masculino',
    'Month-to-month': 'Mensal',
    'One year': 'Anual',
    'Two year': 'Bianual',
    'Electronic check': 'Cheque Eletrônico',
    'Mailed check': 'Cheque por Correio',
    'Bank transfer (automatic)': 'Transferência Bancária (Auto)',
    'Credit card (automatic)': 'Cartão de Crédito (Auto)'
}

# Selecionando colunas que contém texto/categorias
colunas_para_traduzir = df.select_dtypes(include=['category', 'object', 'str']).columns


# Aplicando o dicionário de tradução
for col in colunas_para_traduzir:
    if col != 'ID_Cliente': # Protegendo a coluna de IDs
        df[col] = df[col].astype('object').replace(traducao_geral).astype('category')

print("--- Dataframe Padronizado e 100% em Português ---")
display(df.head())
```
